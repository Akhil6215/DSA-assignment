#Performing meaningful visualizations using the dataset

```python
plt.bar(data['Classification'], data['PW'])
plt.title('Bar graph')
plt.xlabel('Classification')
plt.ylabel('PW')
plt.show()
```

```python
sns.lineplot(x='Classification', y='SW', data=data)
plt.title('Line Plot')
plt.xlabel('Classification')
plt.ylabel('SW')
plt.show()
```

```python
data.plot.box(vert=False, positions=[1, 2, 3, 4])
plt.title('box plot')
```

```python
data.plot(subplots=True, figsize=(8, 8))
```
